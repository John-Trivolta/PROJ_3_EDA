```python
list1 = list()
df = hotels.copy()
for i in df['tags']:
    string = i.replace('[', '').replace(']', '')
    string = string.replace("' ", '').replace(" '", '')
    tags_list = string.split(',')
    tags_list = list(map(lambda x: x.strip(), tags_list)) 
    for j in range(0, len(tags_list)):  
        list1.append(tags_list[j])
```

```python
from collections import Counter
count = Counter(list1)
count.most_common()[0]
```

```python
dict1 = dict()
for key, val in count.items():
    if 'Stayed' in key:
        dict1[key] = val
sorted = dict(sorted(dict1.items(), key=lambda item: item[1]))
```
